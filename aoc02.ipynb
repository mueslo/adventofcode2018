```python
def inv_d(d):
    inv_map = {}
    for k, v in d.items():
        inv_map[v] = inv_map.get(v, [])
        inv_map[v].append(k)
    return inv_map
```

```python
import numpy as np
nums = list(np.array(list(map(ord, [c for c in box_id]))) for box_id in box_ids)

found = False
for i1, s1 in enumerate(box_ids):
    for i2, s2 in enumerate(box_ids[:i1]):
        #diff = abs(nums[i1] - nums[i2])
        diff = (nums[i1] != nums[i2])
        sd = sum(diff)
        #print(i1, i2, sd)
        if sd == 1:
            print('indices', i1, i2)
            print('strings', s1, s2)
            print('common', "".join(np.array(list(s1))[np.logical_not(diff)]))
            found = True
            break
    if found:
        break
print(found)
```
