```python
from itertools import cycle

m = {}
dir_d = {'<': -1, '>': 1, '^': 1j, 'v': -1j}
carts = {}
for i, row in enumerate(input.split("\n")):
    for j, c in enumerate(row):
        loc = j-i*1j
        if c in r'/\+':
            m[loc] = c
        elif c in dir_d:
            carts[loc] = dir_d[c], cycle([1j, 1, -1j])

while len(carts) > 1:
    for loc in sorted(carts, key=lambda x: (-x.imag, x.real)):
        if loc not in carts:
            continue  # deleted due to collision
        dxn, turn = carts.pop(loc)
        loc += dxn
        
        if loc in carts:
            print('collision!', loc, '(cart', loc - dxn)
            del carts[loc]
            continue
        
        track = m.get(loc)
        if track == '+':
            dxn = dxn * next(turn)
        elif track is not None: #/ or \
            dxn *= 1j * (2*((track == '/') ^ (dxn.real == 0))-1)
        carts[loc] = dxn, turn
        
print(len(carts))
print(carts)
```
